```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import classification_report
from sklearn.ensemble import RandomForestClassifier
```

```python
X, y = make_classification(n_classes=2, n_features=25, n_samples = 5000, n_informative=10, n_redundant=5, flip_y=0.1, random_state=
                          42, weights=[0.4, 0.6])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, train_size=0.8)
```

```python
rf_classifier = RandomForestClassifier(random_state=42, verbose=True)
```

```python
rf_classifier.fit(X_train, y_train)
```

```python
y_predict = rf_classifier.predict(X_test)
```

```python
print(classification_report(y_test, y_predict))
```

```python
params_grid = {
    'n_estimators':[50, 100, 150],
    'max_depth':[None,3,5],
    'criterion':['gini', 'entropy'],
    'max_features':['sqrt', 'auto', 'log2'],
}
```

```python
grid_search = GridSearchCV(estimator=rf_classifier, param_grid=params_grid,cv = 5, scoring='accuracy',n_jobs=-1)
```

```python
grid_search.fit(X_train, y_train)
```

```python
best_param = grid_search.get_params
best_param
```

```python
best_model = grid_search.best
y_predict = best_model.predict(X_test)
```

```python
print(classification_report(y_test, y_predict))
```
